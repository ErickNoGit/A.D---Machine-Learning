# **- Mostrar a quantidades de celulares vendidos em touchscreen**

- Quantidade de celulares em relação à marca no rank

```python
#Quantidade de celulares para comparação
dataFrame.Manufacturer.value_counts()
```

No rank de 115 aparelhos mais vendidos, 37 deles são da Samsung, 27 deles da Nokia e 16 da Apple

```python
#Mostrando
grafico = dataFrame.Manufacturer.value_counts()
grafico.plot(kind = 'bar')
```

```python
# especificando apenas a forma do aparelho em toque de tela
NovoData = dataFrame.loc[dataFrame['Form factor'] == 'Touchscreen'].value_counts()
display(NovoData)
```

**-Mostrando a quantidade de celulares em Form factor**

```python
#todos os tipos e seus valores
quantidade = dataFrame['Form factor'].value_counts()
display(quantidade)
```

```python
#porcentagem em relação ao Rank total
porcentagem = (74*100) / 115
print('A quantidade de celulares vendidos em touchscreen é 74, representando %.2f%%' %(porcentagem))
```

# **- Comparando com os menos vendidos**

```python
#Os mais vendidos são tipo 'Bar' e 'Touchscreen'
quantidade.plot(kind='bar')
```

```python
#calculando porcentagem total do grafico
NovaLista = [74, 27, 6, 4, 2, 1, 1]
formLista = ['Touchscreen =', 'Bar =','Slider =','Flipphone =','Keyboardbar =','Taco =','Tiltslider =']
listaVazia = []
j = 0
for i in NovaLista:
  totalGrafico = (i*100) / 115
  listaVazia.append('%.2f%%' %(totalGrafico))
  print(formLista[j], listaVazia[j])
  j += 1
```

```python
#Comparação gráfica
grafico = pd.crosstab(index= dataFrame['Manufacturer'], columns= dataFrame['Form factor'])
grafico.plot(kind='bar')
```

### - Celulares em Touchscreen e em Bar representam juntos 87.83% a mais que os outros tipos de aparelho, sendo essa uma das principais razões na classificação do TOP 3. Sendo Samsung, Nokia e Apple. 

### - Xiaomi e Huawei tentam acompanhar os padrões touchscreen, mas não conseguem por diversos fatores externos. Mesmo assim, apresentam uma quantidade enorme e crescente de vendas, tendo foco nesse modelo.

### - No Rank total a Xiaomi produziu 7 celulares Touchscreen e à Huawei 6 desse comparativo.
